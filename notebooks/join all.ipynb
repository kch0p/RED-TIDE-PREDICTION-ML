```python
buoys = pd.read_csv('lj_torey_buoys.csv')
buoys['datetime'] = pd.to_datetime(buoys['datetime'])
buoys['datetime'] = buoys['datetime'].dt.strftime('%Y-%m-%d')


nutrients = pd.read_csv('nutrients_simplified(NO_UNITS).csv').iloc[:,1:]
nutrients.rename(columns={'time':'datetime'},inplace=True)
nutrients['datetime'] = pd.to_datetime(nutrients['datetime'])
nutrients['datetime'] = nutrients['datetime'].dt.strftime('%Y-%m-%d')


scripps_temp_sal = pd.read_csv('scripps_temp_sal_2008-2021.csv')
scripps_temp_sal['datetime'] = pd.to_datetime(scripps_temp_sal['datetime'])
scripps_temp_sal['datetime'] = scripps_temp_sal['datetime'].dt.strftime('%Y-%m-%d')
```
